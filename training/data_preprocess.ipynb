```python
import shutil
import os
from glob import glob


for folder in range(1,6):
    cnt = 1
    img_list = sorted(glob(f'./data/{folder}/*/*.jpg'))
    for img in img_list:
        shutil.copy(img,f"./data/{folder}/train_{cnt}.jpg")
        cnt += 1
```

```python
import pandas as pd

comic_df = pd.DataFrame(columns=['img_path',1,2,3,4,5])
comic_df.loc[0] = 1,1,1,1,1,1
comic_df
```

```python
cnt = 0
for folder in range(1,6):
    img_list = sorted((glob(f'./data/{folder}/*.jpg')))
    label = [0 for _ in range(6)]
    label[folder-1] = 1
    for img in img_list:
        comic_df.loc[cnt] = img,label[0],label[1],label[2],label[3],label[4]
        cnt += 1
comic_df 
```
