```python
import shutil
import os
from glob import glob
```

```python
# folder 이름 train_1.jpg 로 바꾸기
folder_path = sorted(glob(f'./data/*'))
for f in folder_path:
    cnt = 1
    img_path = sorted(glob(f+'/*.jpg'))
    for img in img_path:
        shutil.move(img,f+f'/train_{cnt}.jpg')
        cnt += 1
```
